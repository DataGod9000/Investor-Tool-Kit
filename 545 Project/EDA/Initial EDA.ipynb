```python
import yfinance as yf
import pandas as pd

# Define the stock ticker and pull data for the last 5 years
ticker = 'AAPL'
stock = yf.Ticker(ticker)

# Get historical market data (daily)
data = stock.history(period='5y')
print(data.head())
```

```python
# Get stock info
stock_info = stock.info
print("Stock Info:", stock_info)

# Get recent financials
income_statement = stock.financials
balance_sheet = stock.balance_sheet
cash_flow = stock.cashflow

print("Income Statement:\n", income_statement)
print("Balance Sheet:\n", balance_sheet)
print("Cash Flow:\n", cash_flow)
```

```python
import matplotlib.pyplot as plt

# Plot the closing prices
data['Close'].plot(title=f"{ticker} Closing Price Over Time")
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```

```python
import yfinance as yf

# List of stock tickers
tickers = ["AAPL", "MSFT", "GOOGL", "AMZN"]

# Pull historical data for multiple stocks
data = yf.download(tickers, start="2020-01-01", end="2023-01-01")

# Display the data
print(data.head())
close_prices = data['Close']
print(close_prices.head())
daily_returns = close_prices.pct_change().dropna()
print(daily_returns.head())
import matplotlib.pyplot as plt

# Plot each stock's closing price
close_prices.plot(title="Historical Closing Prices")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
tech_stocks = pd.read_csv('nasdaq_tech.csv')
```

```python
print(len(tech_stocks))
```

```python
tech_stocks.head()
```

```python
tech_tickers = tech_stocks['Symbol'].tolist()
```

```python
print(len(tech_tickers))
```

```python
bad_tickers = []
for tick in tech_tickers:
    tick = str(tick)
    tick = tick.strip()
    if len(yf.download(tick, start="2023-01-01", end="2024-01-01")) > 0:
        print('yes')
    else:
        print('no')
        bad_tickers.append(tick)
```

```python
print(len(bad_tickers))
```

```python
cleaned_tickers = []
for ticker in tech_tickers:
    ticker = str(ticker)
    ticker = ticker.strip()
    if ticker not in bad_tickers:
        cleaned_tickers.append(ticker)
```

```python
tech_tickers_string = ""
for ticker in cleaned_tickers:
    tech_tickers_string += str(ticker) + " "
```

```python
print(len(cleaned_tickers))
print(tech_tickers_string)
```

```python
tech_data = yf.download(tech_tickers_string, start="2023-01-01", end="2024-01-01")
```
